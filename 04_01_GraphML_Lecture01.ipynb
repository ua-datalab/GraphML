# **Graph Machine Learning**

**Graph Machine Learning (Graph ML)** refers to a set of techniques and algorithms that leverage the graph structure to perform machine learning tasks.

**Graphs** are a natural way to represent relationships and interactions in various kinds of data, including social networks, biological networks, transportation networks, and more.

**Graph ML** aims to analyze and extract insights from data that can be represented as graphs, tackling tasks such as

*   **Node classification**  : The goal of node classification might be to predict the political affiliation (e.g., conservative or liberal) of each individual based on their connections and the known affiliations of some individuals.

*   **Link prediction**      : Consider an e-commerce platform where users and products are represented as nodes in a bipartite graph, with edges representing purchases. The goal of link prediction is to predict new, potential purchases (i.e., links) that might occur between users and products.

*   **Graph classification** : Chemical Compound Analysis - In this scenario, each graph represents the molecular structure of a chemical compound, with atoms as nodes and chemical bonds as edges. The task of graph classification might involve predicting the properties of these compounds (e.g., whether a compound is toxic or non-toxic) based on their molecular graphs.

### **Centrality Measures**: A set of metrics in graph theory that quantify the importance or influence of a vertex within a graph.

*   **Degree centrality** = The degree centrality for a node v is the fraction of nodes it is connected to.
*   **Closeness centrality** = average length of shortest path to this node from all other nodes
*   **Betweenness centrality** = of all the shortest paths between pairs of node, how many have this particular node in them.
*   **Eigenvector centrality** emphasizes nodes connected to other important nodes.
*   **PageRank** (similar to eigenvector centrality but with a damping factor) identifies nodes with significant influence based on the link structure.






```python
# Calculate centrality measures
degree_centrality = nx.degree_centrality(G_karate)
closeness_centrality = nx.closeness_centrality(G_karate)
betweenness_centrality = nx.betweenness_centrality(G_karate)
eigenvector_centrality = nx.eigenvector_centrality(G_karate)
pagerank = nx.pagerank(G_karate)

print(f"{'Node':>4} | {'Degree':>6} | {'Closeness':>9}| {'Betweenness':>11}  | {'Eigenvector':>11}  | {'PageRank':>8}")
print("-" * 60)

# Iterate through each node
for node in G_karate.nodes():
    print(f"{node:4} | {degree_centrality[node]:.4f} | {closeness_centrality[node]:.4f}   | "
          f"{ betweenness_centrality[node]:.4f}       | {eigenvector_centrality[node]:.4f}       | "
          f"{pagerank[node]:.4f}")
```

## **Community Detection**

#### Community detection algorithms try to partition the network into communities or clusters, optimizing some criteria. A common goal is to maximize the number of within-community edges while minimizing the number of between-community edges.

**Community Detection Algorithms**

*   **Modularity Maximization:** Measures the strength of division of a network into modules (or communities). A higher modularity value indicates a stronger division.

  *   **The Louvain method** is a popular greedy optimization method that iteratively groups nodes to maximize modularity.
  *   **Greedy modularity maximization** begins with each node in its own community and repeatedly joins the pair of communities that lead to the largest modularity until no further increase in modularity is possible (a maximum).

*   **Girvan-Newman Algorithm:** Identifies communities by progressively removing edges from the original graph. Edges are removed based on their edge betweenness scores (the number of shortest paths running through an edge).
*   **Label Propagation:** This algorithm assigns labels to each node and then updates them based on the majority labels of their neighbors. It's a fast method and naturally adapts to changes in the network's structure.

## **Graph traversal algorithms**
## **BFS and DFS**


---



**Breadth-First Search (BFS)**

BFS starts at a selected node (root) and explores its neighbors before moving on to the neighbors of those neighbors. This process continues until all nodes are visited or the desired condition is met. BFS is implemented using a queue and is particularly useful for finding the shortest path on unweighted graphs.

**Applications:**


*   Shortest path in unweighted graphs.
*   Crawling websites (web spiders).



---



**Depth-First Search (DFS)**

DFS starts at the root node and explores as far as possible along each branch before backtracking. This algorithm uses a stack (which can be implemented via recursion) to keep track of the nodes being visited. DFS is useful for tasks that need to explore all possible solutions or paths in a graph.

**Applications:**


*   Puzzle and maze solving.
*   Finding connected components.


---



<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/0*vVEol7YdLLG230S-.png" width="500">



The key distinction between DFS and BFS lies in their approach to graph exploration.

* DFS delves deeper into the graph, prioritizing the exploration of nodes as far away from the starting point as possible, following each branch to its end before backtracking.

* In contrast, BFS takes a broader approach, first examining all nodes directly adjacent to the starting node, then progressively moving to nodes further away, exploring the graph in layers based on proximity.

# BFS

```python
def bfs(graph, start_node):
    """
    Perform a breadth-first search (BFS) on a graph from a start node.

    Parameters:
    - graph: A dictionary representing the adjacency list of the graph where keys are node identifiers,
             and values are lists (or sets) of identifiers representing the neighbors of each node.
    - start_node: The node from which the BFS starts.

    - time complexity is O(|V| + |E|)

    Returns:
    - A list of nodes visited during the BFS in the order they were visited.
    """
    # Initialize the visited list with the start_node and a queue with the start_node
    visited = [start_node]
    queue = [start_node]

    while queue:
        # Dequeue a node from the front of the queue
        current_node = queue.pop(0)

        # Explore each neighbor of the current node
        for neighbour in graph[current_node]:
            if neighbour not in visited:
                visited.append(neighbour)  # Mark the neighbour as visited
                queue.append(neighbour)  # Add the neighbour to the queue for further exploration

    return visited
```

```python
def dfs(graph, start_node):
    """
    Perform a depth-first search (DFS) on a graph from a start node.

    Parameters:
    - graph: A dictionary representing the adjacency list of the graph where keys are node identifiers,
             and values are lists (or sets) of identifiers representing the neighbors of each node.
    - start_node: The node from which the DFS starts.

    - time complexity O(|V| + |E|)

    Returns:
    - A list of nodes visited during the DFS in the order they were visited.
    """
    visited = []  # Initialize the visited list

    def dfs_helper(node):
        """Helper function for DFS recursion."""
        if node not in visited:
            visited.append(node)
            for neighbour in graph[node]:
                dfs_helper(neighbour)

    dfs_helper(start_node)
    return visited
```

## **Why Calculating All Paths First (Brute-Force) is Inefficient**

---




### *   **Computational Complexity:** Calculating all possible paths between two nodes in a graph is a highly complex process, especially as the size of the graph increases. The number of simple paths in a graph can grow exponentially with the number of nodes, leading to very high computational costs.


### *   **Memory Usage:** Storing all possible paths can also consume a significant amount of memory, which is inefficient and unnecessary when you only need the shortest path.



## coming back to the **Shortest Path Algorithm** - intuitive understanding of **Dijkstra's Algorithm**


### Imagine you're in a city trying to find the quickest route to a friend's house. You have a map that shows the time it takes to travel each street. Dijkstra's algorithm is like a smart strategy to find the fastest path without wandering around unnecessarily.

###  Here’s how it works:

### **Starting Point:** You begin at your house, marking it as your starting point.

### **Exploration:** From your starting point, you look at all the streets you can directly travel to and note how long each journey would take. You pick the shortest one and move there.

### **Update and Decide:** Now, from this new point, you again look at all the places you can reach directly from here, including your friend's house. If you find a quicker way to a place you've already checked out, you update your notes. Then, you choose the next shortest path from your current location and move forward.

### **Repeat:** You keep repeating this process—looking around from your current spot, updating times, and moving to the next closest location—until you finally reach your friend's house.

### **Result:** The path you've taken, following the shortest travel time at each step, is the quickest route to your friend's house.

### **Dijkstra's algorithm** optimizes the pathfinding process by maintaining a priority queue of nodes to visit based on the current shortest distance.

### This optimization is absent in the brute-force approach, which doesn't prioritize paths based on their current length or weight.

```python
# Use Dijkstra's algorithm to find the shortest path and its length
source='Brighton'
target='Bluevale'
length, path = nx.single_source_dijkstra(G_cities, source=source, target=target)

print(f"Shortest path from {source} to {target}: {path} with length {length}")
```

```python
import random

# Fix the random seed for reproducibility
random.seed(42)

# Initialize the graph
G_large_complex = nx.Graph()

# Generate 100 cities named City0, City1, ..., City99
cities = [f"N{i}" for i in range(100)]

# Add cities to the graph
G_large_complex.add_nodes_from(cities)

# Programmatically create routes between cities
for city in cities:
    connected_cities = random.sample(cities, k=random.randint(2,3))  # Each city will connect to other landmarks
    for connected_city in connected_cities:
        if city != connected_city:  # Ensure no city connects to itself
            # Random weight for the route,
            weight = random.randint(20, 100)
            G_large_complex.add_edge(city, connected_city, weight=weight)

# Example: Inspecting the number of connections and a sample of routes
print(f"Total number of connections (edges): {len(G_large_complex.edges())}")
```

## **Some considerations**


---


#### In summary, while calculating all paths and then selecting the shortest one might seem conceptually simple, it is highly inefficient compared to using Dijkstra's algorithm. Dijkstra's method minimizes computational effort and memory usage by intelligently navigating the graph, making it a superior choice for most shortest pathfinding applications.



---



#### In scenarios modeled like our city distances problem, especially when considering the possibility of shortcuts or other factors that could introduce negative distances, the limitations of Dijkstra's algorithm become evident. It remains an excellent choice for many applications with positive weights but falls short in environments where negative weights come into play, highlighting the importance of choosing the right tool for the problem at hand. **Bellman-Ford**, despite being slower, offers a robust alternative in such cases.



---

## Illustrating Bellman-Ford

## **Delivery Network Among Cities**

### Consider a delivery network connecting the following cities:

*   New York
*   Los Angeles
*   Chicago
*   Houston
*   Phoenix
*   Philadelphia
*   San Antonio

### The delivery costs vary due to factors like distance, demand, and promotional discounts on certain routes.



```python
G_delivery = nx.DiGraph()


# Edges with updated weights to create a scenario where Bellman-Ford shines
G_delivery.add_weighted_edges_from([
    ("New York", "Chicago", 15),
    ("New York", "Houston", 20),
    ("Houston", "San Antonio", 10),
    ("San Antonio", "Chicago", -25),  # Discounted route
    ("Chicago", "Los Angeles", 15),
    ("Houston", "Los Angeles", 30),
    ("San Antonio", "Tucson", -20),   # Another discounted route
    ("Tucson", "Los Angeles", 20),
    ("Atlanta", "Tucson", 25),
    ("Atlanta", "Miami", 10),
    ("Miami", "New York", -15),       # Discount from Miami to New York
    ("Atlanta", "New York", 20),
    ("Tucson", "Chicago", 10),
    ("Los Angeles", "Miami", 45),     # Long route with a higher cost
])

source_city = "New York"
```

```python
#get edge labels
edge_labels = nx.get_edge_attributes(G_delivery, 'weight')
```

```python
plt.figure(figsize=(6, 6))
nx.draw(G_delivery,  with_labels=True, font_color='k', node_size=600, node_color="lightblue",edge_color='#00b4d9',pos=nx.shell_layout(G_delivery))
nx.draw_networkx_edge_labels(G_delivery, pos=nx.shell_layout(G_delivery), edge_labels=edge_labels)
plt.show()
```

## **Applying Bellman-Ford**
### First, we'll use Bellman-Ford to find the **most cost-effective routes** from **New York** to all other cities, accounting for discounts:

```python
try:
    distances, paths = nx.single_source_bellman_ford(G_delivery, source_city)
    print("Cost-effective delivery paths from New York, considering discounts:")
    for city, cost in distances.items():
        print(f"To {city}: Cost = {cost}, Path = {paths[city]}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
try:
    dijkstra_distances = nx.single_source_dijkstra_path_length(G_delivery, source_city)
    print("\nAttempted delivery paths using Dijkstra's algorithm from New York:")
    for city, cost in dijkstra_distances.items():
        print(f"To {city}: Cost = {cost}")
except Exception as e:
    print(f"An error occurred with Dijkstra: {e}")
```
